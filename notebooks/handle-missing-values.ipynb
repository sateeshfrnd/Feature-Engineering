## Imputation Missing Values

### Mean Value Imputation

```python
sns.histplot(df['age'],kde=True)
```

```python
# compute the mean of the available values.
age_mean_value = df['age'].mean().round() # rounded to its nearest integer since we are dealing with age.
age_mean_value
```

```python
# Replace all missing values with the mean
df['age_mean'] = df['age'].fillna(age_mean_value)
```

```python
df[['age_mean','age']]
```

As you can see, the age of the instance with index 888 has changed from NaN to 30, which is the mean that was calculated. The same procedure occurs for all 177 NaN values.

```python
sns.histplot(df['age_mean'],kde=True)
```

*Mean Imputation works well when we have normally distributed data*

### Median Value Imputation
- If we have outliers in the dataset use median

```python
# compute the median of the available values.
age_median_value = df['age'].median() 
age_median_value
```

```python
df['age_median']=df['age'].fillna(age_median_value)
```

```python
df[['age_median','age']]
```

As you can see, the age of the instance with index 888 has changed from NaN to 28, which is the median that was calculated. The same procedure occurs for all 177 NaN values.

```python
df[['age_median','age_mean','age']]
```

### Mode Imputation Technqiue
Deal with Categorical features

Categorical features are those that comprise discrete values typically belonging to a finite set of categories. Categorical data can be nominal or ordinal. Nominal refers to categories that do not follow a specific order, such as music genre or city names, whereas ordinal refers to categories with a sense of order, such as clothing sizes or level of education.

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
mode_value=df[df['embarked'].notna()]['embarked'].mode()[0]
mode_value
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]
```
